```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python
rootUrl = "https://en.wikipedia.org"
mainUrl = "https://en.wikipedia.org/wiki/List_of_sharks"
```

```python
html_doc = urlopen(mainUrl)
soup= bs(html_doc, 'lxml')
```

```python
alphaSort=soup.find('span', {'id' : "Alphabetic_sort"})
alphaSort
```

```python
import pandas as pd
df = []
headings = ["Name", "Status", "Species"]


theList = DSharks.find_all('a')
for item in theList:
    eachUrl = rootUrl + item['href']
    eachPage = urlopen(eachUrl)
    eachSoup = bs(eachPage)
    
    newRow = []
    newRow.append(item['title']) #extract the shark name from title of a ref
    
    biota = eachSoup.find('table', class_ = 'biota') #making the code more efficient. this part of code is common for conStatus and species
    
    # Return the conservations status
    conStatus = biota.find('a', title = 'Conservation status')
    
    if conStatus is not None:
        newRow.append(conStatus.find_parent().find_parent().find_parent().find_next_siblings()[0].find('a')['title'])
    
    else:
        newRow.append("MISSING")
    
    # Return the species
    newRow.append(biota.find('div', class_ = 'species').text)
    
    # Create Data Frame from row list
    df.append(newRow)

print(pd.DataFrame(df,columns = headings))
```
